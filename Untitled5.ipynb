```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
from IPython.display import HTML, clear_output, display
from google.colab.output import eval_js
from base64 import b64decode
import shutil
from google.colab import files

# 🎯 คำศัพท์ใหม่
vocab_list = ['กิน', 'นอน', 'ตื่น', 'สาย']

# สร้างโฟลเดอร์หลัก
dataset_dir = 'thai_speech_dataset'
os.makedirs(dataset_dir, exist_ok=True)

# สร้างโฟลเดอร์ย่อย
for word in vocab_list:
    os.makedirs(os.path.join(dataset_dir, word), exist_ok=True)

print(f"✅ สร้างโฟลเดอร์ {vocab_list} เรียบร้อยแล้ว!")
```

```python
#@title 🎛️ เลือกคำศัพท์ที่กำลังจะอัด
TARGET_WORD = "กิน" #@param ["กิน", "นอน", "ตื่น", "สาย"]

AUDIO_HTML = """
<script>
var my_div = document.createElement("DIV");
var my_btn = document.createElement("BUTTON");

my_btn.style.padding = "15px 30px";
my_btn.style.fontSize = "20px";
my_btn.style.cursor = "pointer";
my_btn.style.backgroundColor = "#d93025";
my_btn.style.color = "white";
my_btn.style.border = "none";
my_btn.style.borderRadius = "8px";

var t = document.createTextNode("🎤 กดเพื่ออัดคำว่า: " + TARGET_WORD_JS);
my_btn.appendChild(t);
my_div.appendChild(my_btn);
document.body.appendChild(my_div);

var base64data = 0;
var reader;
var recorder, gumStream;
var recordButton = my_btn;

var handleSuccess = function(stream) {
  gumStream = stream;
  recorder = new MediaRecorder(stream);

  recorder.ondataavailable = function(e) {
    reader = new FileReader();
    reader.readAsDataURL(e.data);
    reader.onloadend = function() {
      base64data = reader.result;
    }
  };

  recorder.start();
  recordButton.innerText = "🔴 กำลังบันทึก...";

  setTimeout(() => {
    recorder.stop();
    gumStream.getAudioTracks()[0].stop();
    recordButton.innerText = "✅ เสร็จแล้ว!";
  }, 1500);
};

recordButton.onclick = function() {
  navigator.mediaDevices.getUserMedia({audio: true})
    .then(handleSuccess);
};

function getAudio() {
  return new Promise(resolve => {
    var interval = setInterval(() => {
      if (base64data != 0) {
        clearInterval(interval);
        resolve(base64data);
      }
    }, 100);
  });
}
</script>
"""

display(HTML(f"<script>var TARGET_WORD_JS = '{TARGET_WORD}';</script>"))
display(HTML(AUDIO_HTML))

data = eval_js("getAudio()")
binary = b64decode(data.split(',')[1])

# แปลงไฟล์
temp_webm = 'temp.webm'
temp_wav = 'temp.wav'

with open(temp_webm, 'wb') as f:
    f.write(binary)

!ffmpeg -y -i {temp_webm} -ac 1 -ar 16000 {temp_wav} -loglevel quiet

# ปรับขนาดเสียงให้ 1 วินาที
y, sr = librosa.load(temp_wav, sr=16000)
expected_length = 16000

if len(y) > expected_length:
    y = y[:expected_length]
elif len(y) < expected_length:
    y = np.pad(y, (0, expected_length - len(y)), 'constant')

# บันทึกไฟล์
filename = f"{TARGET_WORD}_{int(time.time())}.wav"
save_path = os.path.join(dataset_dir, TARGET_WORD, filename)
sf.write(save_path, y, sr)

clear_output()
print("===========================================")
print(f"✅ บันทึก: {save_path}")
print(f"📏 ขนาด: {len(y)} จุด")
print("===========================================\n")

# แสดงกราฟ
plt.figure(figsize=(10,2))
plt.plot(y)
plt.title(f'Waveform: {TARGET_WORD}')
plt.xlim(0,16000)
plt.axis('off')
plt.show()

!rm -f {temp_webm} {temp_wav}
```

```python
print("📊 สรุป Dataset")

total = 0
for word in vocab_list:
    path = os.path.join(dataset_dir, word)
    count = len(os.listdir(path))
    total += count
    print(f"{word}: {count} ไฟล์")

print("-----------------------")
print(f"รวมทั้งหมด: {total} ไฟล์")
```
